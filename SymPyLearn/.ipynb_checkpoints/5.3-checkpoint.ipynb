```python
from sympy import *
from sympy.plotting import plot3d
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
x,y=symbols('x, y')
```

```python
f1 = x*(y-5)**2
f1
```

```python
plot3d(f1)
```

```python
solve(f1)
```

```python
f2 = (y - exp(1) / 20)**2 * sin(pi * x)
f2
```

```python
plot3d(f2)
```

```python
solve(f2)
```

```python
f3 = x * sin(x)*sin(y - 2)
f3
```

```python
plot3d(f3)
```

```python
solve(f3)
```
