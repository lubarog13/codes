```python
from sympy import *
from sympy.plotting import plot3d
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
a2,a1, a0=symbols('a2, a1, a0')
```

```python
MSE = 1/3 * (((2*a2 + 200*a1 + a0) - 200)**2 +((a2 + 300*a1 + a0) -450)**2 + ((3*a2 + 600*a1 + a0) - 550)**2)
MSE
```

```python
MSEa2 = diff(MSE, a2)
MSEa2
```

```python
MSEa1 = diff(MSE, a1)
MSEa1
```

```python
MSEa0 = diff(MSE, a0)
MSEa0
```

```python
nonlinsolve([MSEa2,MSEa1,MSEa0], [a2,a1,a0])
```
